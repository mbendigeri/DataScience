```python
#This function will group by indexcol and get the count of unique values in each of the column given in reqcolumns
# Refer the examples http://localhost:8888/notebooks/IPython%20Basics/Group%20by%20with%20nunique.ipynb 
#to understand this example
def get_subdataframes(datafr,indexcol,reqcolumns):
    Qltycheckdataframe=datafr.groupby(indexcol).agg({col:"nunique" for col in reqcolumns})
   
    #Qltycheckdataframe=Qltycheckdataframe.drop("refNum",axis=1)
                               
    #Qltycheckdataframe=Qltycheckdataframe.drop("games",axis=1)
    
    
    shapeval= Qltycheckdataframe[Qltycheckdataframe>1].dropna(how="all").shape[0]
    
    print (shapeval)
    if shapeval!=0:
        print("Warning !, The dataframe that you are using is not unique")
    return datafr.groupby(indexcol).agg({col:"max" for col in reqcolumns})
```

```python
players=get_subdataframes(dataframe,player_index,players_cols)
players.head()
```

```python
def save_subgroup(dataframe,index,subgroupname,prefix="raw"):
    save_group_file="_".join([prefix,subgroupname,".csv_gz"])
    print("File name {0}",save_group_file)
    dataframe.to_csv(save_group_file,compression="gzip",encoding="UTF-8")
    test_df=pd.read_csv(save_group_file,compression="gzip",index_col=index,encoding="UTF-8")
    if dataframe.equals(test_df):
        print("Test passed -data is the same as you save")
    else:
        print("Warning -- equivalence test!!! Double-check.")
        
```

```python
save_subgroup(players, player_index, "players")
```
