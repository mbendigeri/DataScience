### do some exploratory analysis to see influence of variables on the price

#### Which bedrooms have more sales

```python
# using factor plot
sns.factorplot(x='bedrooms',data=housing_df,kind='count',size=6)
```

#### Draw another plot in descending order

```python
#using the value_counts().plot
plt.figure(figsize=(8,8))
sns.set(font_scale=1.2)
plt.title('number of Bedroom')
plt.xlabel('Bedrooms')
plt.ylabel('Count')
housing_df['bedrooms'].value_counts().plot(kind='bar')
#sns.despine
#plt.boxplot(housing_df['bedrooms'])
#plt.show()
```

```python
plt.figure(figsize=(16,16))
sns.jointplot(x=housing_df.lat.values, y=housing_df.long.values, size=10)
plt.ylabel('Longitude', fontsize=12)
plt.xlabel('Latitude', fontsize=12)
plt.show()
#sns.despine
```

### which locations has more houses
#### Visualizations the houses based on the location , longitude and latitude

```python
plt.figure(figsize=(18,18))
sns.set_context("notebook", font_scale=1.1)
sns.set_style("ticks")
sns.lmplot('lat','long',data=housing_df,hue='bedrooms',size=8,fit_reg=False)
plt.ylabel('Longitude', fontsize=12)
plt.xlabel('Latitude', fontsize=12)
#plt.show()
```

#### We can see that most are between 47 and 48 latittue and 122.2 and 122.4 lontitude

#### still it is not clear to know which location have more houses . Need to see more better usualization

### Let us see which features are effecting the dependent variable "Price"

#### What is relationship between proce and bedrooms

```python
corr=housing_df.corr()
plt.figure(figsize=(16,16))
sns.set(font_scale=1.2)
sns.heatmap(corr,cmap='coolwarm',xticklabels=corr.columns.values,yticklabels=corr.columns.values,annot=True)
```

```python
sns.regplot(x=housing_df['price'],y=housing_df['bedrooms'])
```

```python
sns.regplot(x=housing_df['price'],y=housing_df['sqft_living'])
```
