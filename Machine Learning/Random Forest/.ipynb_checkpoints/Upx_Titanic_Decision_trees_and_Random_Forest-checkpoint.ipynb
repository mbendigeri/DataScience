```python
your_local_path="C:/Users/tejks/Desktop/ML/practice/"
```

```python
import pandas as pd
from sklearn.preprocessing import Imputer
from sklearn import tree
from sklearn import metrics
import numpy as np
import matplotlib.pyplot as plt
% matplotlib inline
train_df = pd.read_csv(your_local_path+"titanic.csv")
test_df = pd.read_csv(your_local_path+"titanic_test.csv")
```

```python
train_df
```

```python
train_df["Sex"] = train_df["Sex"].apply(lambda sex: 0 if sex == 'male' else 1)
```

```python
train_df.head()
```

```python
y = targets = labels = train_df["Survived"].values

columns = ["Pclass", "Sex", "Age", "SibSp"]
features = train_df[list(columns)].values
features
```

```python
imp = Imputer(missing_values='NaN', strategy='mean', axis=0)
X = imp.fit_transform(features)
X
```

```python
from sklearn.cross_validation import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=123)
```

```python
my_tree_one = tree.DecisionTreeClassifier(criterion="entropy", max_depth=3)
my_tree_one = my_tree_one.fit(X_train, y_train)
```

```python
my_tree_one
```

```python
#The feature_importances_ attribute make it simple to interpret the significance of the predictors you include
print(my_tree_one.feature_importances_) 
print(my_tree_one.score(X_train, y_train))
print(my_tree_one.score(X_test, y_test))
list(zip(columns,my_tree_one.feature_importances_))
```

```python
with open("titanic.dot", 'w') as f:
    f = tree.export_graphviz(my_tree_one, out_file=f, feature_names=columns)
```

#### visit http://webgraphviz.com/  to visualize the tree

```python
pred = my_tree_one.predict(X_test)
pred
```

```python
#Print Confusion matrix 

df_confusion = metrics.confusion_matrix(y_test, pred)
df_confusion
```

```python

def plot_confusion_matrix(df_confusion, title='Confusion matrix', cmap=plt.cm.gray_r):
    plt.matshow(df_confusion, cmap=cmap) # imshow
    plt.title('Confusion Matrix')
    plt.colorbar()
    plt.ylabel('Actual')
    plt.xlabel('Predicted')

plot_confusion_matrix(df_confusion)
```

```python
# ROC curve on Predicted probabilities
pred_proba = my_tree_one.predict_proba(X_test)
fpr, tpr, _ = metrics.roc_curve(y_test, pred_proba[:,1])
roc_auc = metrics.auc(fpr, tpr)
plt.title('Receiver Operating Characteristic')
plt.plot(fpr, tpr, 'b',
label='AUC = %0.2f'% roc_auc)
plt.legend(loc='lower right')
plt.plot([0,1],[0,1],'r--')
plt.xlim([-0.1,1.2])
plt.ylim([-0.1,1.2])
plt.ylabel('True Positive Rate')
plt.xlabel('False Positive Rate')\\
```

#### What happen when we change the tree depth?

```python
#Setting "max_depth" to 10 and "min_samples_split" to 5 : my_tree_two
max_depth = 10
min_samples_split = 5
my_tree_two = tree.DecisionTreeClassifier(max_depth = 10, min_samples_split = 5, random_state = 123)
my_tree_two = my_tree_two.fit(X_train, y_train)

#Print the score of the new decison tree
print(my_tree_two.score(X_test, y_test))
print(my_tree_one.score(X_test,y_test))
```
