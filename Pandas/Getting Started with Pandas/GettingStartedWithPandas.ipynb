## Lists

```python
#Lists
friends = ['Mallu', 'Arvind', 'Tirupati']
friends.append('Veeresh')
```

```python
['Mallu', 'Arvind', 'Tirupati', 'Veeresh']
```

```python
#Check an item
'Somya' in friends
```

```python
#index operations
 friends[0]
friends[-1] #  len(friends) - 1
```

```python
#slicing operations 
friends[1:2]
print(' All friends :', friends)
fr=friends[:1]   # Implicit start at 0
print(' only frirst', fr)
fr=friends[1:]   # Implicit end at len(people)
print(' 1st to end', fr)
fr=friends[::2]  # Take every other item
print('Alternate',fr)
fr=friends[::-1] # Reverse sequence
print('Reverese', fr)
```
