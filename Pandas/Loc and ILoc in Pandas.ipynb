#### Multiple columns and rows can be selected using iloc

```python
# Mutiple  selections using iloc and DataFrame
dataframe.iloc[0:5] #first 5 rows
dataframe.iloc[:,0:3]#first three columns
dataframe.iloc[[0,2,55,66],[0,2,3]] # rows with row numbers 0,2,55,66, columns 0,2,3
# note rows that are selected are from 0 to 4 i,e [x,y] will reult in x to y-1
dataframe.iloc[0:5,5:8]# rows from 0 to 5 and columns from 5 to 8
```

#### .iloc return a series when one row or one column is selected. And returns a dataframe when mutiple rows are selected
#### To get a Dataframe when one row is selected, use a list selection

```python
print(type(dataframe.iloc[100])) # returns a series as only one row is selected
print(type(dataframe.iloc[[100]])) # returns a dataframe because we have used a list selection as only one row is selected
print(type(dataframe.iloc[0:1])) # returns a dataframe as 2 rows are returned
#dataframe.iloc[0:2,0]
print(type(dataframe.iloc[0:2,0])) # returns a Series as we have selected only one column
print(type(dataframe.iloc[0:2,[0]])) # returns a dataframe as we have used a list selection
```

#### Rows and columns selecte are from the first number x , to one minus the second one to y-1

#### Boolean /Logical indexing
##### Pass a series of conditions or true or false values to select a series that satisfy the condition

```python
dataframe.reset_index(inplace=True)
```

```python
dataframe.loc[dataframe['first_name']=='Simona']
```

```python
dataframe.loc[dataframe['first_name']=='Simona','last_name':'email']# returns a dataframe
```

```python
dataframe.loc[dataframe['first_name']=='Simona',['address','email']]# returns a dataframe
```

```python
dataframe.loc[dataframe['first_name']=='Simona','email']# returns a series
```

```python
dataframe.loc[dataframe['email'].str.endswith('gmail.com') & (dataframe['first_name']=='Simona')] 
```

```python
dataframe.loc[dataframe['first_name'].isin(['Simona','Antonia'])] 
```

```python
dataframe.loc[(dataframe['id']>50) & (dataframe['id']<100),['zip','web']]
```
